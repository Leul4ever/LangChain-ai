How to split by character-Character Text Splitter

This is the simplest method. This splits based on a given character sequence, which defaults to "\n\n". Chunk
length is measured by  number of characters.

1. How the text is split: by single character separator.
2. How the chunk size is measured: by number of characters.

```python
from langchain_community.document_loaders import TextLoader
loader= TextLoader("speech.txt")
docs= loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=100, chunk_overlap=20)
text_splitter.split_documents(docs)
```

```python
speech=""
with open("speech.txt") as f:
    speech=f.read()
text_splitter=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
text = text_splitter.split_text(speech)
print(text[0])
print(text[1])

```
